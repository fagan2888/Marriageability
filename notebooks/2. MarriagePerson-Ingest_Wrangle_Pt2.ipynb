```python
#Importing Python packages
import pandas as pd
import numpy as np 
import math
import timeit
```

```python
#Importing ACSdata
ACSmarry = pd.read_csv('data/ACSmarry.csv', sep=',', header=0, skipinitialspace=True)
```

```python
def int_mak(num):  
    split_num = str(num).split('.')
    int_part = int(split_num[0])
    return int_part
```

```python
def int_math(num):  
    num = math.floor(num)
    return num
```

```python
# Job Code Transformation function
#input must be interger data type 

def jobid_to_jobclass(occp):
    try:
#     try: 
        
#         occp = int_math(occp)
#         print(occp)
# #         split_num = str(occp).split('.')
# #         occp = int(split_num[0])
#     except:
#         return 'UNE'
#     else:

        if (occp >= 10 and occp <= 430):
            jobclass = 'MGR'
        elif occp <= 740:
            jobclass = 'BUS'
        elif occp <= 950:
            jobclass = 'FIN'
        elif occp <= 1240:
            jobclass = 'CMM'
        elif occp <= 1560:
            jobclass = 'ENG'
        elif occp <= 1965:
            jobclass = 'SCI'
        elif occp <= 2060:
            jobclass = 'CMS'
        elif occp <= 2160:
            jobclass = 'LGL'
        elif occp <= 2550:
            jobclass = 'EDU'
        elif occp <= 2920:
            jobclass = 'ENT'
        elif occp <= 1240:
            jobclass = 'CMM'
        elif occp <= 3540:
            jobclass = 'MED'
        elif occp <= 3655:
            jobclass = 'HLS'
        elif occp <= 3955:
            jobclass = 'PRT'
        elif occp <= 4150:
            jobclass = 'EAT'
        elif occp <= 4250:
            jobclass = 'CMM'
        elif occp <= 4650:
            jobclass = 'PRS'
        elif occp <= 4965:
            jobclass = 'SAL'
        elif occp <= 5940:
            jobclass = 'OFF'
        elif occp <= 6130:
            jobclass = 'FFF'
        elif occp <= 6765:
            jobclass = 'CON'
        elif occp <= 6940:
            jobclass = 'EXT'
        elif occp <= 7630:
            jobclass = 'RPR'
        elif occp <= 8965:
            jobclass = 'PRD'
        elif occp <= 9750:
            jobclass = 'TRN'
        elif occp <= 9830:
            jobclass = 'MIL'
        elif occp <= 9920:
            jobclass = 'UNE' 
        else:
            jobclass = 'UNE'  
            
        return jobclass
    except:
        return 'Error'
```

```python
#Create List of integers to be converted in 3 Charatacter job code
OCCP_int = []
#start = timeit.timeit()
for i in ACSmarry['OCCP']:
    try:
        OCCP_int.append(int_math(i))
    except: 
         OCCP_int.append(9999)# if Cell is empty assume Unemployed
#end = timeit.timeit()
#print(end - start)
```

```python
#Create list of Job Codes
OCCP_list = []
for i in OCCP_int:
    try:
        OCCP_list.append(jobid_to_jobclass(i))
    except: 
         OCCP_list.append(0)
#print(OCCP_list)
ACSmarry['OCCP_T'] = pd.DataFrame(OCCP_list)
```
